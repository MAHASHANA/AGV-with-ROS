```python
import numpy as np
from typing import Tuple
from tqdm import trange
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
import numpy as np
sns.set_theme(style="whitegrid")

class BanditEnv:
    """Multi-armed bandit environment"""

    def __init__(self, k: int) -> None:
        """__init__.

        Args:
            k (int): number of arms/bandits
        """
        self.k = k

    def reset(self) -> None:
        """Resets the mean payout/reward of each arm.
        This function should be called at least once after __init__()
        """
        # Initialize means of each arm distributed according to standard normal
        self.means = np.random.normal(size=self.k)

    def step(self, action: int) -> Tuple[float, int]:
        """Take one step in env (pull one arm) and observe reward

        Args:
            action (int): index of arm to pull
        """
        # TODO calculate reward of arm given by action
        reward = np.random.normal(loc=self.means[action], scale=1)
        return reward



def q4(k: int, num_samples: int):
    """Q4

    Structure:
        1. Create multi-armed bandit env
        2. Pull each arm `num_samples` times and record the rewards
        3. Plot the rewards (e.g. violinplot, stripplot)

    Args:
        k (int): Number of arms in bandit environment
        num_samples (int): number of samples to take for each arm
    """

    env = BanditEnv(k=k)
    env.reset()

    reward = {}
    for arm in range(k):
        reward[arm+1] = []
        for step in range(num_samples):
            sample = env.step(arm)
            reward[arm+1].append(sample)

    reward_df = pd.DataFrame(data=reward)

    reward_plot = sns.violinplot(data=reward_df, color="gray", inner=None, width=0.8)
    reward_plot.set_ylabel("Reward Distribution")
    reward_plot.set_xlabel("Actions")
    plt.title("Distribution of Sampled Rewards for Each Action")

    plt.show()

q4(10, 1000)
```

```python
from abc import ABC, abstractmethod
from typing import Optional, Sequence


def argmax(arr: Sequence[float]) -> int:
    """Argmax that breaks ties randomly

    Takes in a list of values and returns the index of the item with the highest value, breaking ties randomly.

    Note: np.argmax returns the first index that matches the maximum, so we define this method to use in EpsilonGreedy and UCB agents.
    Args:
        arr: sequence of values
    """
    return np.argmax(arr)


class BanditAgent(ABC):
    def __init__(self, k: int, init: int, step_size: float) -> None:
        """Abstract bandit agent class

        Implements common functions for both epsilon greedy and UCB

        Args:
            k (int): number of arms
            init (init): initial value of Q-values
            step_size (float): step size
        """
        self.k = k
        self.init = init
        self.step_size = step_size

        # Q-values for each arm
        self.Q = None
        # Number of times each arm was pulled
        self.N = None
        # Current total number of steps
        self.t = None

    def reset(self) -> None:
        """Initialize or reset Q-values and counts

        This method should be called after __init__() at least once
        """
        self.Q = self.init * np.ones(self.k, dtype=np.float32)
        self.N = np.zeros(self.k, dtype=int)
        self.t = 0

    @abstractmethod
    def choose_action(self) -> int:
        """Choose which arm to pull"""
        raise NotImplementedError

    @abstractmethod
    def update(self, action: int, reward: float) -> None:
        """Update Q-values and N after observing reward.

        Args:
            action (int): index of pulled arm
            reward (float): reward obtained for pulling arm
        """
        raise NotImplementedError


class EpsilonGreedy(BanditAgent):
    def __init__(self, k: int, init: int, epsilon: float, step_size: Optional[float] = None) -> None:
        """Epsilon greedy bandit agent

        Args:
            k (int): number of arms
            init (init): initial value of Q-values
            epsilon (float): random action probability
            step_size (float or None): step size. If None, then it is equal to 1 / N_t (dynamic step size)
        """
        super().__init__(k, init, step_size)
        self.epsilon = epsilon

    def choose_action(self):
        """Choose which arm to pull

        With probability 1 - epsilon, choose the best action (break ties arbitrarily, use argmax() from above).
        With probability epsilon, choose a random action.
        """
        # TODO
        x = np.random.uniform()
        if x > self.epsilon:
            action = argmax(self.Q)
        else:
            action = np.random.randint(self.k)
        return action

    def update(self, action: int, reward: float) -> None:
        """Update Q-values and N after observing reward.

        Args:
            action (int): index of pulled arm
            reward (float): reward obtained for pulling arm
        """
        self.t += 1

        # TODO update self.N
        self.N[action] += 1

        # TODO update self.Q
        # If step_size is given (static step size)
        if self.step_size is not None:
            self.Q[action] += self.step_size*(reward - self.Q[action])
        # If step_size is dynamic (step_size = 1 / N(a))
        else:
            self.Q[action] += (reward - self.Q[action])/self.N[action]


class UCB(BanditAgent):
    def __init__(self, k: int, init: int, c: float, step_size: float) -> None:
        """Epsilon greedy bandit agent

        Args:
            k (int): number of arms
            init (init): initial value of Q-values
            c (float): UCB constant that controls degree of exploration
            step_size (float): step size (use constant step size in case of UCB)
        """
        super().__init__(k, init, step_size)
        self.c = c

    def choose_action(self):
        """Choose which arm to pull

        Use UCB action selection. Be sure to consider the case when N_t = 0 and break ties randomly (use argmax() from above)
        """
        # TODO
        out = []
        for i in range(self.k):
            q = self.Q[i]
            n = self.N[i]
            if n == 0:
                return i
            else:
                out.append(q + self.c * np.sqrt(np.log(self.t) / n))

        action = argmax(out)
        return action

    def update(self, action: int, reward: float) -> None:
        """Update Q-values and N after observing reward.

        Args:
            action (int): index of pulled arm
            reward (float): reward obtained for pulling arm
        """
        self.t += 1

        # TODO update self.N
        self.N[action] += 1

        # TODO update self.Q
        self.Q[action] += self.step_size * (reward - self.Q[action])
```

```python
def plot_graph(data, axis, color, label, steps):
    avg = []
    for i in data:
        avg.append(np.mean(i))
    axis.plot(avg, color=color, label=label)
    avg = np.array(avg)
    #(1.96 standard error) given in the question
    y_err = 1.96 * avg.std() * np.sqrt(1/len(avg))
    axis.fill_between(np.linspace(0, steps-1, steps), avg-y_err, avg+y_err, alpha=0.2, color=color)


def q6(k: int, trials: int, steps: int):
    """Q6

    Implement epsilon greedy bandit agents with an initial estimate of 0

    Args:
        k (int): number of arms in bandit environment
        trials (int): number of trials
        steps (int): total number of steps for each trial
    """
    # TODO initialize env and agents here
    env = BanditEnv(k=k)
    agents = [EpsilonGreedy(k, 0, 0.1), EpsilonGreedy(k, 0, 0.01), EpsilonGreedy(k, 0, 0)]
    fig, axs = plt.subplots(2, 1)

    # Loop over trials
    rewards0 = np.zeros((steps, trials))
    rewards01 = np.zeros((steps, trials))
    rewards001 = np.zeros((steps, trials))

    optimals0 = np.zeros((steps, trials))
    optimals01 = np.zeros((steps, trials))
    optimals001 = np.zeros((steps, trials))

    best_action = np.zeros(trials)

    for t in trange(trials, desc="Trials"):
        # Reset the environment and agents after every trial
        env.reset()

        optimal = np.argmax(env.means)
        best_action[t] = (env.means[optimal])
        for agent in agents:
            agent.reset()

            for s in range(steps):
                action = agent.choose_action()
                reward = env.step(action)

                if agent.epsilon == 0:
                    rewards0[s][t] = reward

                    if action == optimal:
                        optimals0[s][t] = 1

                elif agent.epsilon == 0.1:
                    rewards01[s][t] = reward

                    if action == optimal:
                        optimals01[s][t] = 1

                elif agent.epsilon == 0.01:
                    rewards001[s][t] = reward

                    if action == optimal:
                        optimals001[s][t] = 1
                agent.update(action, reward)

    # TODO For each trial, perform specified number of steps for each type of agent

    plot_graph(rewards0, axs[0], 'g', 'e=0', steps)
    plot_graph(rewards01, axs[0], 'b', 'e=0.1', steps)
    plot_graph(rewards001, axs[0], 'r', 'e=0.01', steps)

    plot_graph(optimals0, axs[1], 'g', 'e=0', steps)
    plot_graph(optimals01, axs[1], 'b', 'e=0.1', steps)
    plot_graph(optimals001, axs[1], 'r', 'e=0.01', steps)

    axs[0].set_xlabel('Steps')
    axs[1].set_xlabel('Steps')
    axs[0].set_ylabel('Average Reward')
    axs[1].set_ylabel('Optimal Action')

    axs[1].yaxis.set_major_formatter(PercentFormatter(xmax=1, decimals=0))

    upper_bound = np.mean(best_action)
    axs[0].axhline(y=upper_bound, color='k', linestyle='-', label='Upper Bound')
    y_err = 1.96 * best_action.std() * np.sqrt(1 / len(best_action)) #Error as mentioned in the question
    axs[0].fill_between(np.linspace(0, steps), upper_bound - y_err, upper_bound + y_err, alpha=0.2)

    axs[0].legend(loc=4)
    axs[1].legend(loc=4)



def q7(k: int, trials: int, steps: int):
    """Q7

    Compare epsilon greedy bandit agents and UCB agents

    Args:
        k (int): number of arms in bandit environment
        trials (int): number of trials
        steps (int): total number of steps for each trial
    """
    # TODO initialize env and agents here
    env = BanditEnv(k=k)
    agents = [EpsilonGreedy(k, 0, 0, 0.1), EpsilonGreedy(k, 5, 0, 0.1),
              EpsilonGreedy(k, 0, 0.1, 0.1), EpsilonGreedy(k, 5, 0.1, 0.1), UCB(k, 0, 2, 0.1)]
    fig, axs = plt.subplots(2, 1)

    # Loop over trials
    rewards00 = np.zeros((steps, trials))
    rewards50 = np.zeros((steps, trials))
    rewards001 = np.zeros((steps, trials))
    rewards501 = np.zeros((steps, trials))
    rewardsUCB = np.zeros((steps, trials))

      #  -greedy (Q1 = 0,  = 0)
      #  -greedy (Q1 = 5,  = 0)
      #  -greedy (Q1 = 0,  = 0.1)
      #  -greedy (Q1 = 5,  = 0.1)
      #  UCB (c = 2)

    optimals00 = np.zeros((steps, trials))
    optimals50 = np.zeros((steps, trials))
    optimals001 = np.zeros((steps, trials))
    optimals501 = np.zeros((steps, trials))
    optimalsUCB = np.zeros((steps, trials))

    best_action = np.zeros(trials)

    for t in trange(trials, desc="Trials"):
        # Reset environment and agents after every trial
        env.reset()

        optimal = np.argmax(env.means)
        best_action[t] = (env.means[optimal])
        for agent in agents:
            agent.reset()

            for s in range(steps):
                action = agent.choose_action()
                reward = env.step(action)

                if isinstance(agent, EpsilonGreedy) and (agent.epsilon == 0):
                    if agent.init == 0:
                        rewards00[s][t] = reward

                        if action == optimal:
                            optimals00[s][t] = 1

                    elif agent.init == 5:
                        rewards50[s][t] = reward

                        if action == optimal:
                            optimals50[s][t] = 1

                elif isinstance(agent, EpsilonGreedy) and (agent.epsilon == 0.1):
                    if agent.init == 0:
                        rewards001[s][t] = reward

                        if action == optimal:
                            optimals001[s][t] = 1

                    elif agent.init == 5:
                        rewards501[s][t] = reward

                        if action == optimal:
                            optimals501[s][t] = 1
                else:
                    rewardsUCB[s][t] = reward

                    if action == optimal:
                        optimalsUCB[s][t] = 1
                agent.update(action, reward)


    plot_graph(rewards00, axs[1], 'green', 'e=0, Q1=0', steps)
    plot_graph(rewards50, axs[1], 'blue', 'e=0, Q1=5', steps)
    plot_graph(rewards001, axs[1], 'red', 'e=0.1, Q1=0', steps)
    plot_graph(rewards501, axs[1], 'orange', 'e=0.1, Q1=5', steps)
    plot_graph(rewardsUCB, axs[1], 'cyan', 'UCB c=2', steps)

    plot_graph(optimals00, axs[0], 'green', 'e=0, Q1=0', steps)
    plot_graph(optimals50, axs[0], 'blue', 'e=0, Q1=5', steps)
    plot_graph(optimals001, axs[0], 'red', 'e=0.1, Q1=0', steps)
    plot_graph(optimals501, axs[0], 'orange', 'e=0.1, Q1=5', steps)
    plot_graph(optimalsUCB, axs[0], 'cyan', 'UCB c=2', steps)

    axs[1].set_xlabel('Steps')
    axs[1].set_ylabel('Average Reward')
    axs[0].set_xlabel('Steps')
    axs[0].set_ylabel('Optimal Action')

    axs[0].yaxis.set_major_formatter(PercentFormatter(xmax=1, decimals=0))

    upper_bound = np.mean(best_action)
    axs[1].axhline(y=upper_bound, color='black', linestyle='-', label='Upper Bound')
    y_err = 1.96 * best_action.std() * np.sqrt(1 / len(best_action))
    axs[1].fill_between(np.linspace(0, steps), upper_bound - y_err, upper_bound + y_err, alpha=0.2)

    axs[0].legend(loc=4)
    axs[1].legend(loc=4)

    avg1 = []
    avg2 = []
    for i in range(15):
        avg1.append(np.mean(optimals50[i]))
        avg2.append(np.mean(optimalsUCB[i]))
    print('Q1 = 5:\n', avg1)
    print('UCB:\n', avg2)


q6(k=10, trials=2000, steps=2000) #As per given in the question

q7(k=10, trials=2000, steps=1000) #As per given in the question
plt.show()
```
